```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from xgboost.sklearn import XGBRegressor
import ggplot
%matplotlib inline
```

```python
properties = pd.read_csv('../data/properties_2016.csv')
transactions = pd.read_csv('../data/train_2016_v2.csv')
```

```python
properties.rename(columns={
    'fireplaceflag':'has_fireplace',
    'finishedsquarefeet6':'base_area',
    'finishedsquarefeet12':'finished_area',
    'finishedsquarefeet13':'perimiter_area',
    'finishedsquarefeet15':'total_area',
    'finishedsquarefeet50':'finished_f1_area',
    'finishedfloor1squarefeet':'finished_f1_area2',
    'garagecarcnt':'garage_num',
    'garagetotalsqft':'garage_area',
    'lotsizesquarefeet':'lotsize_area',
    'numberofstories':'story_num',
    'poolsizesum':'pool_area',
    'roomcnt':'rooms_num',
    'yardbuildingsqft17':'patio_area',
    'yardbuildingsqft26':'shed_area',
    'taxvaluedollarcnt':'parcel_tax_appraisal_amt',
    'structuretaxvaluedollarcnt':'house_tax_appraisal_amt',
    'taxamount':'property_tax_appraisal_amt',
    'taxdelinquencyflag':'tax_overdue_amt',
    'taxdelinquencyyear':'tax_overdue_year',
    'calculatedbathnbr':'bathroom_area',
    'threequarterbathnbr':'bathroom_3q_num',
    'basementsqft':'basement_area',
    'bedroomcnt':'bedroom_num',
    'bathroomcnt':'bathroom_num',
    'calculatedfinishedsquarefeet':'calculated_home_area',
    'pooltypeid10':'spa_or_hottub',
    'pooltypeid2':'pool_with_spa_or_hottub',
    'pooltypeid7':'pool_without_hottub',
    'fireplacecnt':'fireplace_num',
    'fireplace_flag':'has_fireplace'
},
inplace=True)
```

```python
properties['tax_overdue_amt'] = properties['tax_overdue_amt'].apply(lambda x: 1 if x == 'Y' else 0)
properties['has_fireplace'] = properties['has_fireplace'].apply(lambda x: 1 if x == True else 0)
```

#### Data Cleaning
We're going to try to use a lot the features here, but there are some where the occurrence is so low that they are simply not useful. 

For some features, their absence still contains information about the property. We can address this fairly simply:

There are a few features which don't seem to be immensely useful. 

```python
# Error Finding
properties['f1_discrepancy_area'] = properties['finished_f1_area'] - properties['finished_f1_area2']
properties['f1_discrepancy_bool'] = (properties['f1_discrepancy_area'] != 0).astype(int)
properties['bedroom_pct'] = (properties['bedroom_num']/properties['rooms_num']).replace({np.inf:0})
properties['bathroom_pct'] = (properties['bathroom_num']/properties['rooms_num']).replace({np.inf:0})
properties['room_error'] = ((properties['bedroom_pct'] > 1.0) | (properties['bathroom_pct'] > 1.0)).astype(int)


# Tax Stuff
properties['house_to_property_tax_pct'] = properties['property_tax_appraisal_amt']/properties['house_tax_appraisal_amt']
properties['parcel_tax_to_property_tax_pct'] = properties['parcel_tax_appraisal_amt']/properties['property_tax_appraisal_amt']
properties['parcel_tax_to_house_tax_pct'] = properties['parcel_tax_appraisal_amt']/properties['house_tax_appraisal_amt']
properties['tax_overdue_to_parcel_tax_pct'] = properties['tax_overdue_amt']/properties['parcel_tax_appraisal_amt']
properties['tax_overdue_to_property_tax_pct'] = properties['tax_overdue_amt']/properties['property_tax_appraisal_amt']
properties['tax_overdue_to_house_tax_pct'] = properties['tax_overdue_amt']/properties['house_tax_appraisal_amt']

# Area Calculations
properties['unfinished_area_pct'] = (properties['lotsize_area'] - properties['finished_area'])/properties['lotsize_area']
properties['unfinished_area'] = properties['lotsize_area'] - properties['finished_area']
properties['avg_garage_area'] = properties['garage_area']/properties['garage_num']
properties['null_pct'] = properties.isnull().sum(axis=1)/properties.shape[1]
```

```python
properties.dtypes
```

```python
properties.to_csv('./data/properties_modifiedV2.csv')
```

```python
train = transactions.merge(properties, on='parcelid', how='left')
train['abs_logerror'] = train['logerror'].abs()
```

### Evaluating Feature Importance with XGBoost

```python
THRESH = 0.7
prediction_cols = [c for c in train.columns if c not in ['logerror','abs_logerror','transactiondate','propertyzoningdesc','propertycountylandusecode','parcelid']]
label = 'logerror'
train = train.dropna(thresh=(1-THRESH)*train.shape[0], axis=1)
```

```python
xgb1 = XGBRegressor(
 learning_rate =0.1,
 n_estimators=1000,
 max_depth=5,
 min_child_weight=1,
 gamma=0,
 subsample=0.8,
 colsample_bytree=0.8,
 objective= 'reg:linear',
 eval_metric='mae',
 nthread=4,
 seed=27)
```

```python
xgb1.fit(train[prediction_cols],train[label])
```

```python
a = pd.DataFrame([xgb1.feature_importances_, prediction_cols])
plt.figure(figsize=(12,8))
a.columns = a.iloc[1]
a = a.iloc[0]
a.sort_values().plot(kind='bar')
```

### Setting N_Estimators
To speed up the process, we'll use a slightly higher learning rate (0.1) and set the number of estimators using xgboosts cv and early stopping capabilities.

```python
xgb_param = xgb1.get_params()
xgtrain = xgb.DMatrix(train[prediction_cols].values, label=train[label].values)
cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=1000, nfold=5, metrics='mae', early_stopping_rounds=100)
xgb1.set_params(n_estimators=cvresult.shape[0])
```

### Tuning Part 1
In this section we're tuning the `max_depth` and `min_child_weight` hyperparameters.

```python
from sklearn.model_selection import GridSearchCV
param_test1 = {
    'max_depth':range(3,10,2),
    'min_child_weight':range(1,6,2)
}
gs1 = GridSearchCV(xgb1, param_test1,
                   scoring='neg_mean_absolute_error',
                   iid=False,
                   cv=10                   
                  )

gs1.fit(train[prediction_cols], train[label])
```

```python
print (gs1.best_params_)
results = gs1.grid_scores_
results_df = pd.DataFrame([[r[0]['max_depth'],r[0]['min_child_weight'],r[1]] for r in results],
                         columns=['max_depth','min_child_weight','score'])

sns.heatmap(results_df.pivot(columns='max_depth',index='min_child_weight',values='score'), annot=True)
```

### First Grid Search Results
Not a fantastic score here, but we're still tuning! It looks like we have an optimal `max_depth` of 3 and `min_child_weight` of 5, so we'll fine-tune it a little more and test the values right around each of them.

```python
param_test2 = {
    'max_depth':[2,3,4],
    'min_child_weight':[4,5,6]
}

xgb1.set_params(max_depth=3, min_child_weight=5)

gs2 = GridSearchCV(xgb1, param_test2,
                   scoring='neg_mean_absolute_error',
                   iid=False,
                   cv=10                   
                  )

gs2.fit(train[prediction_cols], train[label])
```

```python
print (gs2.best_params_)
results2 = gs2.grid_scores_
results2_df = pd.DataFrame([[r[0]['max_depth'],r[0]['min_child_weight'],r[1]] for r in results2],
                         columns=['max_depth','min_child_weight','score'])

sns.heatmap(results2_df.pivot(columns='max_depth',index='min_child_weight',values='score'), annot=True)
```

It looks like the combination of `max_depth` = 2 and `min_child_weight` = 6 provided the best scores here. Our model will set those parameters below:

```python
xgb1.set_params(max_depth=2, min_child_weight=6)
```

### Tuning Part 2
Tuning gamma below.

```python
param_test3 = {
    'gamma':[i/10.0 for i in range(5)]
}

gs2 = GridSearchCV(xgb1, param_test3,
                   scoring='neg_mean_absolute_error',
                   iid=False,
                   cv=10                   
                  )

gs2.fit(train[prediction_cols], train[label])
```

```python
results3 = gs2.grid_scores_
results3_df = pd.DataFrame([[r[0],r[1]] for r in results],
                          columns=['gamma','score'])
results3_df.plot()
```

It looks like the optimal value here is just the default - 0. We'll just leave it as-is. 

#### Part 3 - Tuning Tree Values
We'll tune the `subsample` and `colsample_bytree` hyperparameters now.

```python
param_test4 = {
 'subsample':[i/10.0 for i in range(6,10)],
 'colsample_bytree':[i/10.0 for i in range(6,10)]
}

gs4 = GridSearchCV(xgb1, param_test4,
                   scoring='neg_mean_absolute_error',
                   iid=False,
                   cv=10                   
                  )

gs4.fit(train[prediction_cols], train[label])
```

```python
print (gs4.best_params_)
results = gs4.grid_scores_
results_df = pd.DataFrame([[r[0]['subsample'],r[0]['colsample_bytree'],r[1]] for r in results],
                         columns=['subsample','colsample_bytree', 'score'])
sns.heatmap(results_df.pivot(columns='subsample',index='colsample_bytree', values='score'), annot=True)
```

```python
xgb1.set_params(colsample_bytree=0.6, subsample=0.9)
```

It looks like our optimal values found were at 0.6 and 0.9, but there seems to be some room for improvement. We'll do a more thorough search

```python
param_test5 = {
 'subsample':[i/100.0 for i in range(85,101,5)],
 'colsample_bytree':[i/100.0 for i in range(30,70,5)]
}

gs5 = GridSearchCV(xgb1, param_test5,
                   scoring='neg_mean_absolute_error',
                   iid=False,
                   cv=10                   
                  )

gs5.fit(train[prediction_cols], train[label])
```

```python
print (gs5.best_params_)
results = gs5.grid_scores_
results_df = pd.DataFrame([[r[0]['subsample'],r[0]['colsample_bytree'],r[1]] for r in results],
                         columns=['subsample','colsample_bytree', 'score'])
sns.heatmap(results_df.pivot(columns='subsample',index='colsample_bytree', values='score'), annot=True)
```

```python
xgb1.set_params(subsample=1.0, colsample_bytree=0.3)
```

Alright! 

### Part 4 - Additional Regularization
Next up, some of the semi-redundant regularization hyperparameters. I don't expect much improvement here, but it never hurts to try.

```python
param_test6 = {
    'reg_alpha':[0,1e-5,1e-4,1e-2,1,10,100],
    'reg_lambda':[0,.5,1,1.5,2,5,10,100]
}

gs6 = GridSearchCV(xgb1, param_test6,
                   scoring='neg_mean_absolute_error',
                   iid=False,
                   cv=10                   
                  )

gs6.fit(train[prediction_cols], train[label])
```

```python
print (gs6.best_params_)
results = gs6.grid_scores_
results_df = pd.DataFrame([[r[0]['reg_alpha'],r[0]['reg_lambda'],r[1]] for r in results],
                         columns=['reg_alpha','reg_lambda', 'score'])
sns.heatmap(results_df.pivot(columns='reg_alpha',index='reg_lambda', values='score'), annot=True)
```

```python
xgb1.set_params(reg_lambda=5)
param_test7 = {
    'reg_alpha':range(5,20),
}

gs7 = GridSearchCV(xgb1, param_test7,
                   scoring='neg_mean_absolute_error',
                   iid=False,
                   cv=10                   
                  )

gs7.fit(train[prediction_cols], train[label])
```

```python
print (gs7.best_params_)
print (gs7.best_score_)
```
